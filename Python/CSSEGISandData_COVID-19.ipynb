```python
# https://github.com/CSSEGISandData/COVID-19

url_base = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/"
url_ts_confirmed_global = url_base + "time_series_covid19_confirmed_global.csv"
url_ts_deaths_global = url_base + "time_series_covid19_deaths_global.csv"
url_ts_recovered_global = url_base + "time_series_covid19_recovered_global.csv"
```

### **Carregando o dataset**

Os datasets carregados são séries temporais dos casos confirmados, mortes e casos recuperados. Está dividido por países e em alguns casos por províncias, estaremos agrupando as informaçōes com .groupby().


Os datasets ficarão armazenados em um **dict**, verificando o shape e se há valores NaN / não válidos

```python
def read_format_csv(url: str):
    df = pd.read_csv(url, encoding='ascii')
    df_formatted = df.groupby("Country/Region").sum().drop(["Province/State", "Lat", "Long"], axis="columns")
    # del df
    return df_formatted


try:
    confirmed = read_format_csv(url_ts_confirmed_global)
    deaths = read_format_csv(url_ts_deaths_global)
    recovered = read_format_csv(url_ts_recovered_global)
except Exception as error:
    print("\nSome error happened! ", type(error), error)
    exit()


df_dict = {'confirmed': confirmed, 'deaths': deaths, 'recovered': recovered}

FIRST_DATE = df_dict['confirmed'].columns[0]
LAST_DATE = df_dict['confirmed'].columns[-1]  # "3/9/23"

for key, each_df in df_dict.items():
    # print(each_df.shape)
    if not each_df.isnull().sum(axis='columns').sum() and SHOW_INFOS:
        print(f"shape: {each_df.shape}, there is no NaN value in '{key}' dataset!")  # não há NaN

# print('\n', df_dict['confirmed'].head())
# print('\n', df_dict['confirmed'].tail())  # Winter Olympics 2022, some weird things here...
```

### **Limpeza dos datasets**

Estaremos selecionando e ordenando os dados mais recentes (última columa do dataset "confirmed global").

```python
# In this section we are going to verify the most relevant countries, and remove the small ones

df_confirmed_last_tmp = df_dict['confirmed'][LAST_DATE].sort_values(ascending=False)  # getting only the most recent data, last column

df_confirmed_last_tmp.tail(7)
```

Iremos selecionar os países que formam 95% dos casos mundiais, reduzindo para 63 países.

```python
CUT_PERCENT = 0.95
total = df_confirmed_last_tmp.sum()
accum_value = 0
i = 0
for country in df_confirmed_last_tmp.index:
    cut_country = country
    i += 1
    accum_value += df_confirmed_last_tmp.loc[country]
    if accum_value / total > CUT_PERCENT:
        break

print(f"\nThe first {i} countries represent {100 * CUT_PERCENT} % of cases. Previously, we had {df_dict['confirmed'].shape[0]} countries (some of them are not even countries...)")

df_confirmed_last_tmp_2 = df_confirmed_last_tmp.iloc[:i]  # removing countries, slicing
```

```python
sns.barplot(df_confirmed_last_tmp)
plt.show()

sns.barplot(df_confirmed_last_tmp_2)
plt.xticks(rotation=90)
plt.show()
```

Aplicando o corte nos datasets principais/anteriores. Estaremos adicionando uma linha exclusiva representando os casos mundiais "World".

```python
# now we have the most relevant countries (in confirmed cases), we can use this info and reduce / clean other datasets

countries_list = df_confirmed_last_tmp_2.index

for key in df_dict.keys():
    tmp = df_dict.get(key).loc[countries_list]  # removing countries
    tmp.loc['World'] = tmp.sum(axis='index')  # let's add a World entry, sum up all values
    tmp.sort_values(LAST_DATE, ascending=False, inplace=True)
    df_dict[key] = tmp
```

Vamos plotar os dados de alguns países e o total.

```python
# we can plot the time series for each country, or for world


def graf_plot(df: pd.DataFrame, countries: list[str] | tuple[str] | str | int = 0, figsize_t: tuple[float, float] = (12.0, 5.0), title: str = ''):

    if type(countries) is int:
        if countries > 0:
            countries_tmp = []
            for num in range(countries):
                countries_tmp.append(df.index[num])
            countries = countries_tmp
        elif countries <= 0:
            countries = [df.index[0]]
    elif type(countries) is str:
        countries = [countries]

    plt.figure(figsize=figsize_t)
    try:
        for country in countries:
            df.loc[country].plot()
    except TypeError as error:
        raise Exception(f"\n{type(error)}!!! Wrong type in 'countries' argument!!! You must supply a int type or a tuple of strings!")

    plt.title(title)
    plt.legend(countries)
    plt.show()

    return


some_countries = ["World", "US", "India", "Brazil"]
if SHOW_GRAFS:
    graf_plot(df_dict['confirmed'], some_countries, (12, 5), 'COVID-19 - confirmed cases')
    graf_plot(df_dict['deaths'], some_countries, (12, 5), 'COVID-19 - deaths cases')
    graf_plot(df_dict['recovered'], some_countries, (12, 5), 'COVID-19 - recovered')
```

Há um corte de informação no dataset recovered, iremos verificar até qual data vai p/ cada país. Iremos usar essa informação posteriomente.

```python
# the recovered dataset seems to do not have data from some specific date, let's get this date/info for each country,
# we are going to use this info later, when calculating CFR

last_date_r = dict()
tmp_r = df_dict['recovered']
for country in tmp_r.index:
    filtro = tmp_r.loc[country] > 0
    date_array = tmp_r.loc[country][filtro].index  # get the valid columns
    if date_array.size > 0:
        if not last_date_r.get(date_array[-1]):  # if None
            last_date_r[date_array[-1]] = [country]
        else:
            tmp_list = last_date_r.get(date_array[-1])
            tmp_list.append(country)
            last_date_r[date_array[-1]] = tmp_list
        # last_date.add(date_array[-1])  # poderia ser um set(), porém queremos saber os países além da própria data
del tmp_r

for key, value in last_date_r.items():
    print(f'{key}: {value}')

LAST_DATE_RECOVERED = dict()
for key, value in last_date_r.items():  # swapping this list
    if len(value) > 1:
        LAST_DATE_RECOVERED['World'] = key
    else:
        LAST_DATE_RECOVERED[value[0]] = key
```

Os dados nesses datasets são acumulados, podemos calcular e plotar os novos casos diários.

```python
# the datasets we are working are the accumulated data, we can create new datasets for new cases for every day:

for key in tuple(df_dict.keys()):
    df_dict[key + '_diff'] = df_dict.get(key).diff(axis='columns').dropna(axis='columns')

graf_plot(df_dict['confirmed_diff'], some_countries)


# deleting negative value in recovered_diff

for country in df_dict['recovered_diff'].index:
    if country == 'US':
        date_tmp = '12/14/20'  # hardcoded, there's a better way using datetime module
    elif country == 'Serbia':
        date_tmp = '7/20/20'
    elif country == 'Belgium':
        date_tmp = '11/12/20'
    else:
        date_tmp = '8/5/21'
    df_dict['recovered_diff'].loc[country, date_tmp] = 0


graf_plot(df_dict['recovered_diff'].loc[:, :LAST_DATE_RECOVERED['World']], some_countries[1:])

# calculating CFR
```

```python
# Using Fourier to remove noise

def graf_plot_fourier(df: pd.DataFrame, country: str):

    num_samples = df.shape[1]  # 1142
    samples_rate = 7./1.  # 1 sample / 1 day, or 7 samples / 1 week
    # some countries was not counting new cases in the weekends (thus
    # accumulating them in on monday, like Brazil), so we can see
    # in the frequency domain this weekly variation

    F_f = np.fft.rfft(df.loc[country])  # numpy.ndarray
    freq = np.fft.rfftfreq(num_samples, d=1./samples_rate)

    FIG_SIZE = (15, 5)
    if True:
        plt.figure(figsize=FIG_SIZE)
        plt.plot(freq, F_f.real)
        plt.plot(freq, F_f.imag)
        plt.legend(["real", "imag"])
        plt.xlabel('1 / week')
        plt.show()

    freq_cut = 0.5
    noise = F_f.copy()
    # F_f[(freq > 0.8) & (freq < 1.2)] = 0
    F_f[freq > freq_cut] = 0  # removing the noise, 0.12 if sample_rate=1.
    '''
    noise[freq < 0.9] = 0  # valores p/ Brazil
    noise[freq > 1.1] = 0
    '''
    noise[freq < freq_cut] = 0

    low_freq = np.fft.irfft(F_f)  # go back to time domain
    low_freq = pd.Series(low_freq, df.loc[country].index)
    noise_ts = np.fft.irfft(noise)

    plt.figure(figsize=FIG_SIZE)
    low_freq.plot()
    plt.plot(noise_ts)
    # df.loc[country].plot()
    plt.grid()
    plt.show()

    # plt.figure(figsize=FIG_SIZE)
    # plt.plot(noise_ts)
    # plt.legend(['1/week', '2/week', '3/week'])
    # plt.show()

    return

graf_plot_fourier(df_dict['confirmed_diff'], 'US')

# print(df_dict['confirmed_diff'].loc['Brazil'][-200:-150])
```

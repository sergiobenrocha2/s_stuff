# **Testando SymPy - biblioteca p/ matemática simbólica**

"SymPy is a Python library for **symbolic mathematics**. It aims to become a full-featured computer algebra system (CAS)[...]" https://www.sympy.org/en/index.html

O SymPy é uma biblioteca p/ matemática simbólica / computação simbólica (symbolic computation).

Podemos usar no próprio navegador: https://www.sympygamma.com/

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
# import sympy
from sympy import symbols, limit, log, plotting

# print(f'NumPy: {np.__version__}\nSymPy: {sympy.__version__}\n')
```

### **Usando SymPy / testando a função limit():**

```python
# Usando SymPy / testando a função limit():
x, y = symbols('x y')
y = x * log(x)  # aqui ele entendeu que 'y' representa a função, não atribuimos nenhum número (float, int etc...)

print(f'\nlimit(x * log(x), x, 0), ou seja, limit(x * log(x)) com x --> 0+ : {limit(x * log(x), x, 0)}\n')  # aqui calculamos o limite, exemplo retirado do site do SymPy
```

### **Usando SymPy / plotando os simbolos criados c/ a função plotting.plot():**

```python
# Usando SymPy / plotando os simbolos criados c/ a função plotting.plot()
style.use('seaborn-v0_8-whitegrid')
plotting.plot(y, xlim=(0.0, 2.001), ylim=(-0.5, 1.5), axis_center=(0, -0.50))  # podemos plotar também, aqui plotamos o simbolo y

# Comparar com WolframAlpha: https://www.wolframalpha.com/input?i=limit%28x*log%28x%29%2C+x%2C+0%29
```

### **Comparando a resposta c/ NumPy**

```python
# Usando NumPy:
x_ = np.arange(0.001, 2.0, 0.001)  # se x_ < 0, teremos y_ = NaN
y_ = x_ * np.log(x_)

plt.plot(x_, y_)
plt.grid(True)
plt.show()
```

### **Podemos usar também os recursos built-in do Python...**

...p/ plotar e comparar com o resultado do SymPy:



```python
# Usando math, built-in range() e list comprehension:
x_2 = [i/1000.0 for i in range(1, 2 * 1000, 1)]
y_2 = [each_x * math.log(each_x) for each_x in x_2]

plt.plot(x_2, y_2)
plt.grid(True)
plt.show()
```
